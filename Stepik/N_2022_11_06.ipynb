```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
students_perfomance=pd.read_csv("https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv",encoding="windows 1251")
students_perfomance.head()
```

```python
students_perfomance["math score"].hist()
```

```python
students_perfomance.plot.scatter(x="math score", y="reading score")
```

```python
ax=sns.lmplot(x="math score", y="reading score", hue = "gender", data=students_perfomance, fit_reg=False)
ax.set_xlabels("Math Score")
ax.set_ylabels("Reading Score")
```
